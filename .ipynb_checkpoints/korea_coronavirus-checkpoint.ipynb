# *CLEANING DATA*

## `korean_patient` dataframe:

### Define: Drop `global_num`, `disease` , `infection_order` columns

```python
korean_patient = korean_patient.drop(columns=['global_num', 'infection_order', 'disease'])
```

### Define: Convert `symptom_onset_date`, `confirmed_date`, `released_date`, `deceased_date` columns to datetime. 
### Drop NaN from the`confirmed_date` column 

```python
korean_patient_copy = korean_patient.copy()
```

```python
korean_patient_copy=korean_patient_copy.dropna(subset=['symptom_onset_date', 'released_date'])
```

```python
korean_patient_copy[korean_patient_copy.symptom_onset_date.str.len() !=10]
```

```python
korean_patient.confirmed_date = pd.to_datetime(korean_patient.confirmed_date)
korean_patient.released_date = pd.to_datetime(korean_patient.released_date)
korean_patient.deceased_date = pd.to_datetime(korean_patient.deceased_date)
```

```python
korean_patient = korean_patient.drop([836])
```

```python
korean_patient.symptom_onset_date = pd.to_datetime(korean_patient.symptom_onset_date)
```

```python
korean_patient = korean_patient.dropna(subset=['confirmed_date'])
```

### Define: Convert `age` column from string to int and drop missing values.   

```python
#Take only the number from the string
korean_patient.age = korean_patient.age.str.extract('(\d+)')
```

```python
korean_patient = korean_patient.dropna(subset=['age'])
```

```python
#Convert number in string format to int
korean_patient.age = korean_patient.age.astype(int)
```

### Define: Convert `patient_id` and `infected_by` columns from int to string.   

```python
korean_patient.patient_id = korean_patient.patient_id.astype(str)
korean_patient.infected_by = korean_patient.infected_by.astype(str)
korean_patient.infected_by = korean_patient.infected_by.map(lambda x: x[0:10])
```

```python
korean_patient.head()
```

### Define: Drop NaN values from the `infection_case`, `city` and `sex` columns.   

```python
korean_patient = korean_patient.dropna(subset=['city', 'sex', 'infection_case'])
```

### Define: Create `exposed_days`, `days_noexposure_death`, `days_noexposure_released` columns

```python
korean_patient['exposed_days'] = (korean_patient['confirmed_date']-korean_patient['symptom_onset_date']).dt.days
korean_patient['days_noexposure_death'] = (korean_patient['deceased_date']-korean_patient['confirmed_date']).dt.days
korean_patient['days_noexposure_released'] = (korean_patient['released_date']-korean_patient['confirmed_date']).dt.days
```

### Define: Convert `birth_year`, to `age_number` column

```python
korean_patient['age_number'] = 2020 - korean_patient['birth_year']
korean_patient = korean_patient.dropna(subset=['age_number'])
```

## *Univariate Exploration*

```python
def count_plot(column):
    
    """This function gives receives a column as an argument an then plots a graph of the count of
    unique values of that column."""
    
    base_color = sns.color_palette()[0]
    ko_order = df[column].value_counts().index[:5]
    plt.xticks(rotation=45)
    sns.countplot(data = df, x = column, color = base_color, order = ko_order);
```

### Which are the top 5 cities with more infected people?

```python
count_plot('city')
```

<span style=font-size:1.2em>Seongnam-si is the city that has the most infected people<span>

### Total number of infected people vs time

```python
df_time_series = df.groupby('confirmed_date').patient_id.count().reset_index()
df_time_series['cum_sum'] = df_time_series.patient_id.cumsum()
plt.plot(df_time_series['confirmed_date'], df_time_series['cum_sum'])
plt.xticks(rotation=50)
plt.axvline(pd.Timestamp('2020-02-21'), color='red')
plt.axvline(pd.Timestamp('2020-04-5'), color='red')
plt.tight_layout();
```

<span style=font-size:1.2em>Approximately on the 2020-02-21, the number of total cases starts increasing linearly until 2020-04-21. After that, the rate of change slows down.<span>

## *Multivariate Exploration*

### Histogram age for each state divided by sex

```python
plt.figure(figsize = [30, 30])
g = sns.FacetGrid(data=df, height=3, col='state', col_wrap=1, hue='sex', aspect=2)
g.map(plt.hist, 'age_number', alpha=0.4).add_legend()
plt.show()
plt.tight_layout();
```

<span style=font-size:1.2em>It can be seen that more men than women died. And women who died, died at older age. At a older age,more women than men are being isolated.<span>

### Boxplots showing age range for each sex divided by state

```python
plt.figure(figsize = [30, 30])
g = sns.FacetGrid(data=df, height=3, col='sex', col_wrap=1, aspect=2)
g.map(sns.boxplot,'state','age_number').add_legend()
plt.show()
plt.tight_layout();
```

It can be seen here that the age range of infected male people who has died is larger than of the infected women. Also the median is lower for men than for women.

### What is the average age of males and females for each province?

```python
plt.figure(figsize = [20, 5])
cat_means1 = df.groupby(['state', 'province']).mean()['age_number']
cat_means1 = cat_means1.reset_index(name='mean_age_number')
cat_means1 = cat_means1.pivot(index='state', columns='province', values='mean_age_number')
sns.heatmap(cat_means1, annot = True, fmt = '.1f',
           cbar_kws = {'label' : 'mean_age_number'}, annot_kws={"size": 20}, cmap=sns.cm.rocket_r)
plt.figure(figsize = [20, 5])
cat_means2 = df.groupby(['state', 'province']).count()['patient_id']
cat_means2 = cat_means2.reset_index(name='count')
cat_means2 = cat_means2.pivot(index='state', columns='province', values='count')
sns.heatmap(cat_means2, annot = True, fmt = '.1f',
           cbar_kws = {'label' : 'count'}, annot_kws={"size": 20}, cmap=sns.cm.rocket_r)
plt.figure(figsize = [20, 5])
cat_means3 = df.groupby(['sex', 'province']).count()['patient_id']
cat_means3 = cat_means3.reset_index(name='count')
cat_means3 = cat_means3.pivot(index='sex', columns='province', values='count')
sns.heatmap(cat_means3, annot = True, fmt = '.1f',
           cbar_kws = {'label' : 'count'}, annot_kws={"size": 20}, cmap=sns.cm.rocket_r);
```

<span style=font-size:1.2em>Busan is the 5th province with most cases. There are 63 infected women and 64 infected men. 2 people died in that province and the dead patients were on average 76 years old. Busan province has shown also to have released most of their infected people (110 people). Busan on my perspective has done a good job.<span>

<span style=font-size:1.2em>Gyeonggi-do is the province with the most infected cases. There are 358 infected women and 298 infected men. Only one person has died and the patient was 36 years old. We had to look more into that patient because he was very young to have died of coronavirus. Furthermore, only 58 people have been released and there are 597 people on the wait to be known whether they are going to be released or die. The average age of the isolated people is 44 years old which means that not many people will die.<span>

<span style=font-size:1.2em>Gyeongsangbuk-do is the third province with most cases. There are 196 infected women and 137 infected men. There are 17 dead people which age on average were 72.2 years old. 91 people were released so far. 225 people are on a isolated state. People who are in isolation are 53.1 years old on average.<span>

<span style=font-size:1.2em>Seoul is the second province with more cases. However, there are only 2 deaths. The average age of the dead people were 67.5 years old. ### Average age for each state divided by sex

```python
plt.figure(figsize = [10, 5])
cat_means = df.groupby(['sex', 'state']).mean()['age_number']
cat_means = cat_means.reset_index(name='mean_age_number')
cat_means = cat_means.pivot(index='sex', columns='state', values='mean_age_number')
sns.heatmap(cat_means, annot = True, fmt = '.1f',
           cbar_kws = {'label' : 'mean_age_number'}, annot_kws={"size": 20}, cmap=sns.cm.rocket_r);
```
