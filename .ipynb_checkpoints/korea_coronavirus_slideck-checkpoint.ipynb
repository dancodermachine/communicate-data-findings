# South-Korea Coronavirus Patients

## Investigation Overview

SARS-CoV2 has infected more than 10,000 people in South Korea. The Korea Centers for Disease Control & Prevention (KCDC) organization shares structured data of COVID-19. The aim of this project is to get insights about the data in order to see where to allocate resources to fight coronavirus.


## Dataset Overview

The data consist of approximately 3326 patients. There are 18 characteristics of each patient. Some characteshistics include the three states of a patient, as well as characteristics such as sex, age, etc. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
df = pd.read_csv('korea_coronavirus.csv')
df.head()
```

```python
df.infected_by = df.infected_by.astype(str)
df.infected_by = df.infected_by.map(lambda x: x[0:10])
```

```python
def count_plot(column):
    """This function gives receives a column as an argument an then plots a graph of the count of
    unique values of that column."""
    base_color = sns.color_palette()[0]
    ko_order = df[column].value_counts().index[:5]
    plt.xticks(rotation=45)
    sns.countplot(data = df, x = column, color = base_color, order = ko_order)
    
df.confirmed_date = pd.to_datetime(df.confirmed_date)
```

## 1. Which are the top 5 ways of getting infected?

Most people got infected by being in contact with someone else (872 people). The second source of infection is people who got infected in other country (475 people). The third source is 'others' or 'etc' (422 people). After that, people got infected by attending places where concentration of people exists (e.g. call centers, churhces). This shows that the best way to fight coronavirus is by avoiding contact with others and public places. Furthermore, anyone coming from outside should be place on quarantine since that is the second source of infection in the country. 

```python
a = []
df1 = df.copy()
df1 = df1[df1['infection_case']!='etc']
base_color_1 = sns.color_palette()[0]
ko_order_1 = df1['infection_case'].value_counts().index[:5]
plt.xticks(rotation=45)
sns.countplot(data = df1, x = 'infection_case', color = base_color_1, order = ko_order_1)
for i, v in enumerate(df1['infection_case'].value_counts()[:5]):
    a.append(v)
    plt.text(i-0.15, v-v*0.5, str(v), color='black')   
plt.title('Top 5 ways of getting infected');
```

## 2. Up to what number can one person infect others?

One person can infect up to 51 people!

```python
d = []
base_color = sns.color_palette()[0]
ko_order = df['infected_by'].value_counts().index[1:5]
plt.xticks(rotation=45)
sns.countplot(data = df, x = 'infected_by', color = base_color, order = ko_order)
for i, v in enumerate(df['infected_by'].value_counts()[1:5]):
    d.append(v)
    plt.text(i-0.05, v-v*0.5, str(v), color='white') 
plt.title('Top 5 people who infected others');
```

## 3. Which are the top 5 provinces with more infected people?

The province with most infected people is Gyeonggi-do (656 people). The second province is Seoul with 563 infected people. The third province is Gyenongsangbuk-do with 333 infected people. The fourth province is Chungcheongnam-do with 141 infected people. The fifth province is Busan with 127 infected people. 

```python
c = []
for i, v in enumerate(df['province'].value_counts()[:5]):
    c.append(v)
    plt.text(i-0.15, v-v*0.5, str(v), color='white')   
count_plot('province')
plt.title('Top 5 cities with infected people');
```

## 4. Average age for each state divided by sex

On average, women infected with coronavirus die older (81 years old). However, men on average die younger if they have coronavirus (66.5 years old). And even though there are more women infected than men, more men has died. The reason is that men with coronavirus die younger.

```python
plt.figure(figsize = [10, 5])
cat_means = df.groupby(['sex', 'state']).mean()['age_number']
cat_means = cat_means.reset_index(name='mean_age_number')
cat_means = cat_means.pivot(index='sex', columns='state', values='mean_age_number')
sns.heatmap(cat_means, annot = True, fmt = '.1f',
           cbar_kws = {'label' : 'mean_age_number'}, annot_kws={"size": 20}, cmap=sns.cm.rocket_r)
plt.figure(figsize = [10, 5])
cat_means = df.groupby(['sex', 'state']).count()['patient_id']
cat_means = cat_means.reset_index(name='number_of_cases')
cat_means = cat_means.pivot(index='sex', columns='state', values='number_of_cases')
sns.heatmap(cat_means, annot = True, fmt = '.1f',
           cbar_kws = {'label' : 'number_of_cases'}, annot_kws={"size": 20}, cmap=sns.cm.rocket_r);
```

The average age for the isolated people are 38.8 years old. 

Ulsan has only one dead person which age was 67 years old.
